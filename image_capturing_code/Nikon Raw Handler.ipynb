```python
import numpy as np
import os
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
import rawpy
from PIL import Image 
```

```python
names = [f for f in os.listdir("NEF Captures/") if f.endswith(".NEF")]
names
```

```python
for name in names:
    raw = rawpy.imread("NEF Captures\\" + name)
    rgb = raw.postprocess()
    img = Image.fromarray(rgb)
    img.save("NEF Processed\\" + name.split('.NEF')[0] +".png", format="png") 
```

```python
img.show()
```

```python
coordinates = np.array([[1128.2207, 5358.38], [604.22656, 391.15845], [6374.993, 371.94476], [6073.0977, 5384.163 ]])
```

```python
frameCol = cv.imread("Homography11.JPG")
frame = cv.cvtColor(frameCol, cv.COLOR_BGR2GRAY)
plt.imshow(frame)
```

```python
h, w = frameCol.shape[:2]
```

```python
plane = np.array([[0, 400], [0, 0], [400, 0], [400, 400]])
```

```python
H, mask = cv.findHomography(coordinates, plane)
```

```python
warped = cv.warpPerspective(frameCol, H, (h, w))
plt.imshow(warped)
```
